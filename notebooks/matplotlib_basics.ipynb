```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scipy.stats as st
```

```python
x = np.arange(-3, 3, 0.01)
plt.plot(x, st.norm.pdf(x))
plt.show()
```

```python
plt.plot(x, st.norm.pdf(x))
plt.plot(x, st.norm.pdf(x, -1, 0.5))
plt.savefig("test.png", format='png')
```

```python
axes = plt.axes()
axes.set_xlim([-6,7])
axes.set_ylim([0, 1.1])
axes.set_xticks(np.arange(-6,7,1))
axes.set_yticks(np.arange(0,1.1,0.1))
axes.grid()
plt.plot(x, st.norm.pdf(x))
plt.plot(x, st.norm.pdf(x,-1.0, 0.5))
plt.show()
```

```python
axes = plt.axes()
axes.set_xlim([-6,7])
axes.set_ylim([0, 1.1])
axes.set_xticks(np.arange(-6,7,1))
axes.set_yticks(np.arange(0,1.1,0.1))
axes.grid()
plt.xlabel("Rain")
plt.ylabel("Prob. of Rain XD")
plt.plot(x, st.norm.pdf(x), "b-", label="Patrick")
plt.plot(x, st.norm.pdf(x,-1.0, 0.5), "g-.", label="Amanda")
plt.legend(loc="upper right")
plt.show()
```
