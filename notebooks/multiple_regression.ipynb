```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
df = pd.read_csv("../data/car.csv")
```

```python
df
```

```python
df1 = df[["Mileage", "Price"]]
df1
```

```python
bins = np.arange(0,50000,10000)  # creates an arry thats is like [0,50000) with array[n+1] - array[n] = 10k
```

```python
groups = df1.groupby(pd.cut(df1["Mileage"], bins), observed=True).mean()  # this will give us a new group with range lol
groups.head()
```

```python
groups["Price"].plot.line()
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
X = df[["Mileage", "Age ", "Cylinders"]]
X
```

```python
X_scaled = pd.DataFrame(scale.fit_transform(X), columns=X.columns) #scales all the data between 0 and 1 roughly
X_scaled.tail()
```

```python
y = df[["Price"]]
y.head()
```

```python
X = sm.add_constant(X_scaled) #add columns full of 1 in the beginning of the data frame
X
```

```python
#estimating the regression lol
est = sm.OLS(y,X).fit()
est.summary()
```

```python
y.groupby(df.Cylinders).mean()
```

```python
#testing data uysing the regression result
scaled = scale.transform([[50000, 2, 12]])
scaled = np.insert(scaled[0], 0, 1)
newprice = est.predict(scaled)
print(f"The predicted price of car with specifications (mil, age, # cylinders) 50k, 2 and 12 was found to be {math.ceil(newprice)} dollars")
```
