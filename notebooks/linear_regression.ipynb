```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
pageLoadSpeed = np.random.normal(6.0,5.0,1000)
purchaseAmount = 100 - (pageLoadSpeed + np.random.normal(0,0.1,1000)) * 3
```

```python
plt.scatter(pageLoadSpeed, purchaseAmount)
```

```python
slope, intercept, r_value, p_value, std_error = st.linregress(pageLoadSpeed, purchaseAmount)
```

```python
coeffDetermine = r_value **2
coeffDetermine
```

```python
def line(x_value):
    return slope * x_value + intercept
fitline = line(pageLoadSpeed)
plt.plot(pageLoadSpeed, fitline, c='r')
plt.scatter(pageLoadSpeed, purchaseAmount)
plt.show()
```
