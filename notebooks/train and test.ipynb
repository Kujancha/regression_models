```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
```

```python
#less create a model for predicting the number of visitors on a zoo based on  the tempetruate of a city

temperature = np.random.normal(25.5,5,100)
visitors = np.random.normal(10000,900,100) / (temperature + 0.625)
plt.scatter(temperature, visitors)
```

Now we split the training data into two sets on train set and test set(4:1) to avoid overfitting 

```python
# splitting the data in 4:1 ration
trainX = temperature[:80]
testX = temperature[80:]

trainY = visitors[:80]
testY = visitors[80:]

```

```python
plt.scatter(trainX, trainY)
plt.title("Train Set")
```

```python
plt.scatter(testX, testY)
plt.title("Test set")
```

```python
# let us fit a very high degree polynomial to the train data

poly = np.poly1d(np.polyfit(trainX, trainY,4))   # no. of degrees
poly
```

```python
# create a test set to sketch the curve
datatesting = np.linspace(0,45, 1000)

axes = plt.axes()
axes.set_xlim([0, 45])
axes.set_ylim([0, 1500])
plt.scatter(trainX,trainY)
plt.plot(datatesting, poly(datatesting))
plt.show()
```

```python
# for train set
axes = plt.axes()
axes.set_xlim([0, 45])
axes.set_ylim([0, 1500])
plt.scatter(testX,testY)
plt.plot(datatesting, poly(datatesting))
plt.show()
```

Testing R-squared on both training and test set

```python
from sklearn.metrics import r2_score
r2 = r2_score(trainY, poly(trainX))
r2
```

```python
testr2 = r2_score(testY, poly(testX))
testr2
```
